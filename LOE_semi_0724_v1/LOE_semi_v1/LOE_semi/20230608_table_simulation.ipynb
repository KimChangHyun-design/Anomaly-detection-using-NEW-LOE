```python
path_list =[]
for i in range(len(all_case)):
    simu_num= all_case[i][0]
    contamination=all_case[i][2]
    anomaly_rate =all_case[i][3]
    result_folder = "/media/usr/SSD/jitae/Thesis/Neural transformation/RESULTS/time_series_simulation/simulation_data_ntl_plus/"
    exp_path = os.path.join(f'Simulation_model_{simu_num}_c_{contamination}_a_{anomaly_rate}_{all_case[i][1]}')
    path_list.append(exp_path)
```

```python
import json

path_value_dict = {}   
for i in range(len(path_list)):
    with open(result_folder+path_list[i]+'/assessment_results.json', 'r') as f:
        json_data = json.load(f)

    avg_pc=round(json_data['avg_TS_pc_0'],3)
    std_pc=round(json_data['std_TS_pc_0'],3)
    avg_pf=round(json_data['avg_TS_pf_0'],3)
    std_pf=round(json_data['std_TS_pf_0'],3)
    path_value_dict[i]=path_list[i],avg_pc,std_pc,avg_pf,std_pf 
```
